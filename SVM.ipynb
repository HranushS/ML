```python
np.random.seed(0)


X=np.random.poisson(100,size = (500,2))
y=np.concatenate((np.zeros(500),np.ones(500)))
X1=np.random.poisson(50,size = (500,2))
X=np.concatenate((X,X1))
y=y.astype(int)
```

```python
from sklearn.utils import shuffle
X, y = shuffle(X, y, random_state=0)
```

```python
X[:10],y[:10]
```

```python
counter = Counter(y)
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))

plt.legend()
plt.show()
```

## SVC for linear Kernel
#### Hard Margin

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X,y, test_size=0.2, random_state=10)
```

```python
xTrain.shape,yTrain.shape
```

```python
from sklearn import svm
from sklearn.svm import SVC
```

```python
clf = svm.SVC(kernel="linear",C=10000)
clf.fit(xTrain, yTrain)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(xTrain, yTrain, clf=clf, legend=2)
plt.show()
```

### Scores

```python
clf.score(xTest,yTest)
```

```python
#X, y = make_blobs(n_samples=[500,500], centers=None, n_features=2,random_state=0)
```

```python
X=np.random.poisson(100,size = (500,2))
y=np.concatenate((np.zeros(500),np.ones(500)))
X1=np.random.poisson(75,size = (500,2))
X=np.concatenate((X,X1))
y=y.astype(int)
```

```python
X, y = shuffle(X, y, random_state=0)
```

```python
X.shape, y.shape
```

```python
counter = Counter(y)

for label, _ in counter.items():
	row_ix = where(y == label)[0]
	plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X,y, test_size=0.2, random_state=10)
```

```python
clf = svm.SVC(kernel="linear",C=5)
clf.fit(xTrain, yTrain)
```

```python
plot_decision_regions(xTrain, yTrain, clf=clf, legend=2)
plt.show()
```

```python
clf.score(xTest,yTest)
```

#### RBF Kernel

```python
clf = svm.SVC(kernel="rbf",C=5)
clf.fit(xTrain, yTrain)
```

```python
plot_decision_regions(xTrain, yTrain, clf=clf, legend=2)
plt.show()
```

```python
clf.score(xTest,yTest)
```
